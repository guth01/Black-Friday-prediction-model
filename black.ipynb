```python
df_test=df[df['Purchase'].isnull()]
```

```python
df_train=df[~df['Purchase'].isnull()]
```

```python
y=df_train['Purchase']
```

```python
x=df_train.drop('Purchase',axis=1)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x,y, test_size=0.33, random_state=42)
```

```python
X_train.drop('Product_ID',axis=1,inplace=True)
X_test.drop('Product_ID',axis=1,inplace=True)
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
regressor=RandomForestRegressor()
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_features': [None, 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, 40, 50, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=param_dist,
            cv=5, n_iter=10,
            scoring = 'neg_mean_absolute_error',n_jobs = 3,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
random_cv.fit(X_train,y_train)
```

```python
import xgboost
from sklearn.model_selection import RandomizedSearchCV
regressor=xgboost.XGBRegressor()
booster=['gbtree','gblinear']
base_score=[0.25,0.5,0.75,1]
## Hyper Parameter Optimization


n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
booster=['gbtree']
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }
# Set up the random search with 4-fold cross validation
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=10,
            scoring = 'neg_mean_absolute_error',n_jobs = 3,
            verbose = 5, 
            return_train_score = True,
            random_state=42)
random_cv.fit(X_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error

# Step 1: Hyperparameter tuning using RandomizedSearchCV

# Define the hyperparameter grid
hyperparameter_grid = {
    'fit_intercept': [True, False],  # Whether to include an intercept term
    'copy_X': [True, False],         # Whether to copy X or modify it in place
    'positive': [True, False]       # Whether to normalize the input data
}

# Initialize Linear Regression model
regressor = LinearRegression()

# Set up RandomizedSearchCV
random_cv = RandomizedSearchCV(estimator=regressor,
                               param_distributions=hyperparameter_grid,
                               cv=5, n_iter=5,  # Test 5 random combinations
                               scoring='neg_mean_absolute_error',
                               n_jobs=-1, verbose=5,
                               random_state=42)

# Perform the hyperparameter search on the training data
random_cv.fit(X_train, y_train)

# Step 2: Get the best parameters and train the final model

# Extract the best parameters
best_params = random_cv.best_params_
print("Best Parameters:", best_params)

# Initialize Linear Regression model with the best parameters
best_model = LinearRegression(**best_params)

# Train the model with the entire training dataset
best_model.fit(X_train, y_train)

# Step 3: Make predictions and evaluate

# Predict on the test set
y_pred = best_model.predict(X_test)

# Evaluate the model performance
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```
